# Matplotlib

## Importing Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
#  x = Generating 11 points from 0 to 5 
#  y = (x)^2
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### Functional Method for ploting the

```python
plt.plot(x,y,'g')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
```

#### Sub-plots with functional Method

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object Oriented Method for Plotting 

```python
# creating a figure object
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  #add_axes(leftpadding, bottom padding, width, height)
axes2 = fig.add_axes([0.15,0.5,0.2,0.2])
axes3 = fig.add_axes([0.66,0.2,0.2,0.2])

axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'g')
axes2.set_title('Smaller Plot Top')

axes3.plot(x,x,'b')
axes3.set_title('Smaller Plot Bottom')
```

#### Sub-plots with Object Oriented Method 

```python
# axes is basically a list of matplotlib 
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) # 1 row and 2 cloumn => 2 tables in 1 row 
axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x, 'g')
axes[1].set_title('Second Plot')

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])


ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label = 'X Cube')

ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.legend()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols= 1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
